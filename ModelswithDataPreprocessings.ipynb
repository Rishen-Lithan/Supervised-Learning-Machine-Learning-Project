```python
# Pre-process the data set
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']
```

```python
# Encode species labels to numeric values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Split the Data set
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)
```

```python
# Random Forest Machine Learning Algorithm
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)
```

```python
# Predict & Evaluate
rf_predictions = rf_model.predict(x_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_accuracy_percent = round(rf_accuracy * 100, 2)
```

```python
print("Random Forest Accuracy:", rf_accuracy_percent)
```

```python
# SVM Classifier
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Model Train
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(x_train_scaled, y_train)
```

```python
# Predict & Evaluate
svm_predictions = svm_model.predict(x_test_scaled)
svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_accuracy_percent = round(svm_accuracy * 100, 2)
```

```python
print("Support Vector Machine Accuracy:", svm_accuracy_percent)
```

```python
# MLP Classifier
mlp_model = MLPClassifier(max_iter=1000, random_state=42)
mlp_model.fit(x_train_scaled, y_train)
```

```python
# Predict & Evaluate
mlp_predictions = mlp_model.predict(x_test_scaled)
mlp_accuracy = accuracy_score(y_test, mlp_predictions)
mlp_accuracy_percent = round(mlp_accuracy * 100, 2)
```

```python
print("MLP Classifier Accuracy:", mlp_accuracy_percent)
```

```python
# Plot the Models
models = ['Random Forest', 'SVM', 'MLP']
accuracies = [rf_accuracy, svm_accuracy, mlp_accuracy]
```

```python
plt.bar(models, accuracies, color=['green', 'blue', 'purple'])
plt.ylim(0.9, 1.0)
plt.title('Model Accuracy Comparison on Iris Dataset')
plt.ylabel('Accuracy')
plt.show()
```
